```python
#config PATH
from os import chdir,getcwd
from platform import system
directory = getcwd()
if system()=='Linux':
    directory_list = directory.split('/')
    chdir('/'.join(directory_list[0:len(directory_list)-1])+'/src')
elif system()=='Windows':
    directory_list = directory.split('\\')
    chdir('\\'.join(directory_list[0:len(directory_list)-1])+'\\src')
else:
    raise OSError('system not suported')
```

## Inicialización del paquete, las distribuciones y algunas funciones auxiliares

```python
import sim_2024
sim_2024.set_seed(14501)
from matplotlib import pyplot
#%matplotlib ipympl
```

## Obtención de la muestra

```python
sample_1 = sim_2024.rand(1000)
```

## Visualización

```python
try:
    fig_1 = sim_2024.HistogramFigure(sample_1,bins=[0+1*k/10 for k in range(11)])
    fig_1.axes[0].set_xticks([0.0+k*0.2 for k in range(6)])
    fig_1.axes[0].set_ylim(0,1.3)
    fig_1.axes[0].set_xlim(-0.05,1.05)
finally:
    fig_1
```

```python
f_den = sim_2024.density_function(lambda x:1,min=0.0,max=1.0)
try:
    fig_2 = sim_2024.HistogramFigure(sample_1,f_den,bins=[0+1*k/10 for k in range(11)])
    fig_2.axes[0].set_xticks([0.0+k*0.2 for k in range(6)])
    fig_2.axes[0].set_ylim(0,1.3)
    fig_2.axes[0].set_xlim(-0.05,1.05)
finally:
    fig_2
```

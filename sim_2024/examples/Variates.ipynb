## Distribución Bernoulli

```python
B = Bernoulli(p=0.3)
sample_Bernoulli = B.sample(1000)
try:
    fig_Bernoulli = HistogramFigure(sample_Bernoulli,bins=[-0.5,0.5,1.5],label='Muestra simulada de una distribución tipo Bernoulli con $p=0.3$')
    fig_Bernoulli.axes[0].set_xticks([0,1])
    fig_Bernoulli.axes[0].set_ylim(0,1)
    fig_Bernoulli.axes[0].set_xlim(-0.5,1.5)
finally:
    fig_Bernoulli
```

## Distribución Binomial

```python
Bn = Binomial(p=0.6,n=10)
sample_binomial = Bn.sample(1000)
try:
    fig_binomial = HistogramFigure(sample_binomial,bins=[-0.5+k for k in range(11)],label='Muestra simulada de una distribución binomial con $p=0.3$ y $n=10$')
    fig_binomial.axes[0].set_xticks([k for k in range(11)])
    fig_binomial.axes[0].set_ylim(0,1)
    fig_binomial.axes[0].set_xlim(-0.5,10.5)
finally:
    fig_binomial
```

## Distribución geométrica

```python
G = Geometric(p=0.2)
sample_geometric = G.sample(1000)
try:
    fig_geometric = HistogramFigure(sample_geometric,bins=[-0.5+k for k in range(11)],label='Muestra simulada de una distribución geométrica con $p=0.2$')
    fig_geometric.axes[0].set_xticks([k for k in range(11)])
    fig_geometric.axes[0].set_ylim(0,1)
    fig_geometric.axes[0].set_xlim(-0.5,10.5)
finally:
    fig_geometric
```
